## **LASSO REGRESSION**


In the ridge regression we took alpha for regularization $L=\sum_{i=1}^n (Y'-Y_{i})^2 +\lambda ||w||^{2}$ \
But in the lasso regression we use L1 norm for the regualrization.

Lasso regression is,\
$L=MSE+\lambda ||W||$, where ||W|| is $\left[|W_1|+|W_2|+|W_3|+...+|W_{n}|\right]$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import lasso_path
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_targets=1,noise=20,random_state=13)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
plt.scatter(X,y)
from sklearn.linear_model import LinearRegression,Lasso
lr=LinearRegression()
lr.fit(X_train,y_train)
print(lr.coef_)
print(lr.intercept_)
```

```python
alphas=[0,1,5,10,30]
plt.scatter(X,y)
for i in alphas:
    L=Lasso(alpha=i)
    L.fit(X_train,y_train)
    plt.plot(X_test,L.predict(X_test),label=f'alpha:{i}')
    
plt.legend()
plt.show()
```

Here we see that as the value of alpha is increased the coefficients drops to zero which means coefficients of the columns which are less important will drop to zero. Thus it can be used for the feature selection.
