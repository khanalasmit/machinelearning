```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/kmeans/student_clustering.csv')
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df,x='cgpa',y='iq')
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i)
    kmean.fit(df)
    wcss.append(kmean.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.axvline(x=4,ymin=0,ymax=30000,linestyle='--',color='red',label='elbow point')
plt.legend()
```

```python
X=df.iloc[:,:].values
km=KMeans(n_clusters=4)
y=km.fit_predict(X)
```

```python
y
```

```python
sns.scatterplot(x=df['cgpa'], y=df['iq'], hue=y, palette='viridis', s=100)
# Plot cluster centers
centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=20, linewidths=3, label='Centroids')

plt.title('K-Means Clustering Results (k=4)')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.legend()
plt.show()
```
