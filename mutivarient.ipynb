<p style='font-size:30px'> <b> MULTIVARIENT ANALYSIS </b></p>

<b> In this multivarient analysis is done. Bivarient analysis is the analysis taking two variable for the EDA. If more than two are taken then it is called multivaient analysis. </b>

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

LOADING THE DATASETS FROM THE SNS GIVEN EXAMPLE DATAS

```python
tips=sns.load_dataset('tips')
```

```python
tips.sample(5)
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

<p style='font-size:25px'><b>1. SCATTER-PLOT </b></p>

scatter plot of the tips dataset

here the multivarient analysis is done. from the tips dataset, realtion between total bills and the tip is shown then the color variation is done for sex and the style of the marker is the changed to depict the smoker and non smoker

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', marker='o', s=30,hue='sex',style='smoker',size='size')
```

```python
sns.scatterplot(data=iris,x='petal_length',y='sepal_length',s=30,hue='species')
```

<p style='font-size:25px'><b>2.. BAR PLOT </b></p>

```python
flights.sample(5)
```

```python
sns.barplot(data=flights, x='month', y='passengers', hue='month')
```

```python
sns.barplot(data=flights, x='year', y='passengers', hue='month')

```

<p style='font-size:25px'><b>3.. Box PLOT </b></p>

```python
sns.boxplot(data=tips, x='time', y='total_bill',hue='sex')
plt.show()
```

<p style='font-size:25px'><b>4.. KDE PLOT </b></p>

```python
sns.kdeplot(tips[tips['sex'] == 'Male']['tip'], label='Male', shade=False)
sns.kdeplot(tips[tips['sex'] == 'Female']['tip'], label='Female', shade=False)
plt.legend()
plt.show()
```

<p style='font-size:25px'><b>5.. CONTOUR PLOT </b></p>

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.clustermap(pd.crosstab(titanic['sibsp'],titanic['survived']))
```

<p style='font-size:25px'><b>6.. PAIR PLOT </b></p>

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
sns.pairplot(iris)
```

<p style='font-size:25px'><b>7.. LINE PLOT </b></p>

```python
new1=flights.groupby('year')
total_passengers_per_year = new1['passengers'].sum().plot(kind='line')
print(total_passengers_per_year)
```

```python
new=flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(new)
```

```python
sns.clustermap(new)
```

<p style='font-size:25px'><b>SUB-PLOTS </b></p>

```python
titanic.head()
```

```python
plt.subplot(2, 2, 1)
sns.countplot(data=titanic, x='survived', palette='Set1')
plt.title('Count of the survived and death')

plt.subplot(2, 2, 2)
sex_counts = titanic['sex'].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%')
plt.title('Pie chart of the sex')

plt.subplot(2, 2, 3)
plt.hist(titanic['fare'], bins=20)
plt.title('Histogram to show distribution of fare')

plt.subplot(2, 2, 4)
sns.kdeplot(data=titanic, x='fare', hue='sex')
plt.title('KDE to show the distribution')
plt.suptitle('univarient analysis on the titanic dataset')
plt.tight_layout()
plt.show()
```

```python
iris.sample(5)
```

```python
plt.figure(figsize=(12, 10))  # Adjust the width and height as needed

plt.subplot(2, 2, 1)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')

plt.subplot(2, 2, 2)
sns.scatterplot(data=iris, x='petal_length', y='sepal_length', hue='species', size='petal_width')

plt.subplot(2, 2, 3)
sns.boxplot(data=iris, x='species', y='petal_length')

plt.subplot(2, 2, 4)
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time',style='smoker')  # Note: Pairplot creates its own figure, so it won't fit into the subplot grid.
plt.suptitle('multivareint analysis on iris and tips dataset')
plt.tight_layout()
plt.show()
```
