<b> In this multivarient analysis is done. Bivarient analysis is the analysis taking two variable for the EDA. If more than two are taken then it is called multivaient analysis. </b>

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

LOADING THE DATASETS FROM THE SNS GIVEN EXAMPLE DATAS

```python
tips=sns.load_dataset('tips')
```

```python
tips.sample(5)
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

<p style='font-size:25px'><b>1. SCATTER-PLOT </b></p>

scatter plot of the tips dataset

here the multivarient analysis is done. from the tips dataset, realtion between total bills and the tip is shown then the color variation is done for sex and the style of the marker is the changed to depict the smoker and non smoker

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', marker='o', s=30,hue='sex',style='smoker',size='size')
```

```python
sns.scatterplot(data=iris,x='petal_length',y='sepal_length',s=30,hue='species')
```
