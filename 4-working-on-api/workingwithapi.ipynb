* <p style='font-size:20px'><b>API DEMO : IN THIS I HAVE FETCHED THE API FROM NASA WHICH REQUIRE NO LOGIN OR NO ANY OTHER AUTHENTICATION . YOU CAN JUST FETCH THE DATA AS YOU NEED</b></p>

```python
url='https://api.nasa.gov/DONKI/CME?startDate=2023-01-01&endDate=2024-01-06&api_key=DEMO_KEY'
import requests
import pandas as pd
```

```python
req=requests.get(url)
data=pd.json_normalize(req.json()[0:])
pd.DataFrame(data)
```

```python
pd.DataFrame(data).drop(columns=['catalog','sourceLocation','activeRegionNum','link','versionId'])
```
