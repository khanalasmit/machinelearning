### **DBSCAN**

Density-Based Clustering is a type of unsupervised machine learning technique that groups together data points that are closely packed (dense regions), while marking points in low-density regions as outliers or noise. This approach is particularly useful for finding clusters of arbitrary shapes and handling noisy datasets.

**Key Concepts:**
- *Density:* The number of data points within a certain distance (neighborhood) of a given point.
- *Core Point:* A point that has at least a minimum number of other points (MinPts) within a specified radius (ε or epsilon).
- *Border Point:* A point that is within the neighborhood of a core point but does not itself have enough nearby points to be a core point.
- *Noise (Outlier):* A point that is neither a core point nor a border point.
- *Cluster Formation:* Clusters are formed by connecting core points that are within ε of each other and including their reachable border points.

**Popular Algorithm:**
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is the most well-known density-based clustering algorithm.

**Advantages:**
* Can find arbitrarily shaped clusters.
* Handles noise (outliers) well.
* Does not require the number of clusters to be specified beforehand.

**Disadvantages:**
* Struggles with datasets of varying density.
* Sensitive to the choice of ε and MinPts parameters(hyper-parameters).
