# Notes on Linear Regression

> **Linear regression** is a foundational algorithm in machine learning, known for its simplicity and effectiveness. Mastering this algorithm is essential for understanding more advanced machine learning concepts.

---

## Key Points

- **Linear regression** is typically the first algorithm introduced in machine learning courses, providing a strong foundation for future learning.
- Machine learning can be categorized into:
    - **Supervised learning** (where linear regression is used for regression problems)
    - **Unsupervised learning**
- **Simple linear regression** involves one input variable, while **multiple linear regression** uses several input variables.

---

## Simple vs. Multiple Linear Regression

- **Simple linear regression** predicts outcomes based on a single variable and serves as a basic model for understanding data relationships.
- **Multiple linear regression** extends this to multiple input variables for more complex predictions.

---

## Mathematical Relationship

The mathematical equation for simple linear regression is:

$$
y = \beta_0 + \beta_1 x + \epsilon
$$

Where:

- $( y )$: predicted value (dependent variable)
- $( x )$: input variable (independent variable)
- $\beta_{0} $: intercept
- $\beta_{1}$: slope (shows how changes in \( x \) affect \( y \))
- $\epsilon$: error term

For multiple linear regression:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n + \epsilon
$$

---

- Accurate and well-prepared data is crucial for reliable predictions.
- Real-world data often contains complexities that simple models may not address, highlighting the need for careful data analysis and preparation.

## Drawing the Best-Fit Line

- The goal of linear regression is to draw a line that best fits the data points, minimizing prediction errors.
- This skill is foundational for understanding more complex mathematical and analytical concepts.

---

1. **Prepare and organize data** before applying machine learning techniques.
2. **Split data** into training and testing sets to evaluate model performance.
3. **Visualize relationships** using regression lines to better understand how variables affect outcomes.

---

## Practical Applications

- Linear regression is widely used for data analysis, trend identification, and forecasting.
- It helps in making informed decisions based on predictive models.

---

## Factors Affecting Predictions

- The **slope** (\( \beta_1 \)) indicates the strength of the relationship between variables (e.g., how CGPA affects salary).
- **Experience** and other factors can also impact outcomes, even if the primary variable (like CGPA) is low.

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.iloc[:,[0]]
Y=data.iloc[:,1]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
#plotting the predictec value on the original value
plt.scatter(x=x_test,y=y_test,label='original')
plt.plot(x_test,y_pred,color='orange',label='predicted')
plt.xlabel('CGPA')
plt.ylabel('package')
plt.legend()
```
