* <b> Analysis of the single variable of a data `i.e`  of a single column in a dataset is called univarient analysis </b>

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('C:/Users/Acer/OneDrive/Documents/MACLEARNING/machinelearning/data/advertising.csv')
```

```python
data.sample(5)
```

```python
data['Clicked on Ad'].unique()
```

<p style='font-size:25px'><b>1.Categorical Data</b></p>

<b> a.COUNTPLOT </b>

```python
ax = sns.countplot(x='Clicked on Ad', data=data, palette='Set2')

for p in ax.patches:
    height=p.get_height()
    ax.text(
        p.get_x()+p.get_width()/2,
        height,
        f'{height:.0f}',
        ha='center',
        va='bottom',
        fontsize=10
        
    )

plt.show()
```

```python
bx=sns.countplot(x='Male',data=data,palette='Set3')
for p in bx.patches:
    height=p.get_height()
    bx.text(
        p.get_x()+p.get_width()/2,
        height,
        f'{height:.0f}',
        ha='center',
        va='bottom',
        fontsize=10
        
    )
plt.show()
```

<b>b. PIECHART </b>

```python
data['Clicked on Ad'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
data['Male'].value_counts().plot(kind='pie',autopct='%.2f')
```
