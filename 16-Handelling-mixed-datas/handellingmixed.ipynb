### **HANDELLING MIXED DATA**
Mixed data would be like a column with both numeric and non numeric value.\
In such case we separte the numerical and non numerical values in two different columns and put nan for numeric in non numeric column and vice-versa


```python
import pandas as pd
import numpy as np
```

```python
url='https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/a315efd557d2fc96984ded8f7f440af0320e6ae2/day33-handling-mixed-variables/titanic.csv'
data=pd.read_csv(url)
```

```python
data.head()
```

```python
data['number'].unique()
```

```python
fig=data['number'].value_counts().plot(kind='barh')
```

```python
#extract the numrical part
data['number_numerical']=pd.to_numeric(data['number'],errors='coerce',downcast='integer')
```

```python
data['number_categorical']=np.where(data['number_numerical'].isnull(),data['number'],np.nan)
```

```python
data.head()
```

```python
data['Cabin'].unique()
```

```python
data['cabin_num']=data['Cabin'].str.extract('(\d+)')
data['cabin_cat']=data['Cabin'].str[0]
```

```python
data.head()
```

```python
data['Ticket'].unique()
```

```python
data['ticket_num']=data['Ticket'].apply(lambda s:s.split()[-1])
data['ticket_num']=pd.to_numeric(data['ticket_num'],errors='coerce',downcast='integer')
data['ticket_cat']=data['Ticket'].apply(lambda s:s.split()[0])
data['ticket_cat']=np.where(data['ticket_cat'].str.isdigit(),np.nan,data['ticket_cat'])
```
