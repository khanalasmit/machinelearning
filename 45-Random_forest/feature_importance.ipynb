## **FEATURE IMPORTANCE**
- Finding the importance of the columns of the data

```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/Acer/OneDrive/Documents/MACLEARNING/machinelearning/data/train.csv')
```

```python
#MNIST dataset
df.head()
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
sns.heatmap(X.iloc[5].values.reshape(28,28))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(X,y)
```

```python
rf.feature_importances_
```

```python
rf.feature_importances_.shape
```

```python
#for each rows it has gave the feature importance
#now plotting the heatmap which pixel are important
sns.heatmap(rf.feature_importances_.reshape(28,28))
'''Pixel of 28 rows and 28 columns in a frame'''
```
