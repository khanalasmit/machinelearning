## **MATH BEHIND FEATURE IMPORTANCE**

Feature importance:
- The importance of a feature is computed as the (normalized)total reduction of the criterion brought by that feature. It is also known as gini importance.

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
```

```python
X,y=make_classification(n_samples=5,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=0)#making the datasets
```

```python
clf=DecisionTreeClassifier()
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

**Let's see what is done to find the feature importance on the basis of decision tree**
- we calculate the n

$$
n_i=\frac{N_t}{N}\left[\text{impurity}-\left(\frac{N_{t_r}}{N_t}*\text{right impurity}\right)-\left(\frac{N_{t_l}}{N_t}*\text{left impurity}\right)\right]\\
f_k^{i}=\frac{\sum{n_i}}{\sum{n_i}}
$$

- This we see for the decision tree classifier 
- Random forest uses the multiple decision tree 
- Thus it will take the average of the decision tress 
- Thus value of the random forest may be different

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=2)
rf.fit(X,y)
```

```python
rf.estimators_
```

```python
rf.feature_importances_
```
