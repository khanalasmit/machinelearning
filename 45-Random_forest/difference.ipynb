### **Differences between bagging and random forest**

|**Bagging**|**Random forest**|
|-|-|
|Can use any model as base model|Can use only decision tree classifier as base models|
|during column sampling we use only the selected columns|during column sampling we obtain ramdomness i.e we do the sampling after the formation of nodes|


```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
print(df.shape)
df.head()
```

```python
bag=BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])

```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf=RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[0])
plt.show()
```

WE can see in the above tree that at each node we are selecting different columns for classification thus we are introducing randomness thus we will have better classification.
