$$
\hat{Y}=\sigma (XW)
$$
$$
L=\frac{-1}{ m} \left[\sum_{i=1}^m Y_ilog(\hat{Y_i})+\sum_{i=1}^m(1-Y_i)log(1-\hat{Y_i})\right]
$$


$$
\textbf{Logistic Regression Loss:}
$$

$$
L = -\frac{1}{m} \sum_{i=1}^m \left[ Y_i \log(\hat{Y}_i) + (1 - Y_i) \log(1 - \hat{Y}_i) \right]
$$

where

$$
\hat{Y}_i = \sigma(Z_i) = \frac{1}{1 + e^{-Z_i}}, \quad Z_i = X_i W
$$

---

$$
\textbf{Derivative of Loss with respect to } W:
$$

First, compute the derivative for a single sample:

$$
\frac{\partial L_i}{\partial W} = \frac{\partial}{\partial W} \left[ -Y_i \log(\hat{Y}_i) - (1-Y_i)\log(1-\hat{Y}_i) \right]
$$

$Since ( \hat{Y}_i = \sigma(Z_i) ) and ( Z_i = X_i W ):$

$$
\frac{\partial L_i}{\partial W} = -Y_i \frac{1}{\hat{Y}_i} \frac{\partial \hat{Y}_i}{\partial W} - (1-Y_i) \frac{1}{1-\hat{Y}_i} \left(-\frac{\partial \hat{Y}_i}{\partial W}\right)
$$

$$
= \left( -\frac{Y_i}{\hat{Y}_i} + \frac{1-Y_i}{1-\hat{Y}_i} \right) \frac{\partial \hat{Y}_i}{\partial W}
$$

The derivative of the sigmoid function:

$$
\frac{\partial \hat{Y}_i}{\partial W} = \hat{Y}_i (1 - \hat{Y}_i) X_i
$$

Plug this in:

$$
\frac{\partial L_i}{\partial W} = \left( -\frac{Y_i}{\hat{Y}_i} + \frac{1-Y_i}{1-\hat{Y}_i} \right) \hat{Y}_i (1 - \hat{Y}_i) X_i
$$

Simplify the coefficient:

$$
-\frac{Y_i}{\hat{Y}_i} \hat{Y}_i (1 - \hat{Y}_i) + \frac{1-Y_i}{1-\hat{Y}_i} \hat{Y}_i (1 - \hat{Y}_i) = -Y_i (1 - \hat{Y}_i) + (1-Y_i) \hat{Y}_i
$$

$$
= -Y_i + Y_i \hat{Y}_i + \hat{Y}_i - Y_i \hat{Y}_i = \hat{Y}_i - Y_i
$$

So,

$$
\frac{\partial L_i}{\partial W} = (\hat{Y}_i - Y_i) X_i
$$

For all samples (vectorized):

$$
\frac{\partial L}{\partial W} = \frac{1}{m} X^T (\hat{Y} - Y)
$$

---

$$
\textbf{Gradient Descent Update Rule:}
$$

$$
W := W - \alpha \frac{\partial L}{\partial W}
$$

or

$$
W := W - \alpha \left( \frac{1}{m} X^T (\hat{Y} - Y) \right)
$$

where \( \alpha \) is the learning rate.
