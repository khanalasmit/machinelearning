### **NAIVE BAYES**

- We calculate the probablity according to the given data.
- The case which has the highere probablity will be chosen as the outcome
- This uses the Bayes theorem to calculate the value of the probablity.
- Here we donot need to calcualt the denominator of the probablity in the Bayes theorem since we will compare two values with same denominator.
- It is called Naives Bayes theorem because it does the basic assumptions. 

# Naïve Bayes Theorem — Mathematical Intuition

1. **Bayes Theorem**  

Bayes theorem relates the conditional probability of an event as:

$$
P(A|B) = \frac{P(B|A)\,P(A)}{P(B)}
$$

where:
- \(P(A|B)\) = Posterior probability (probability of \(A\) given \(B\))  
- \(P(B|A)\) = Likelihood  
- \(P(A)\) = Prior probability of \(A\)  
- \(P(B)\) = Evidence (normalizing constant)

---

2. **Classification Setup**

Suppose we want to classify an instance \(X\) into one of the classes \(C_1, C_2, \dots, C_k\).  

We compute the posterior probability for each class:

$$
P(C_i | X) = \frac{P(X | C_i) \, P(C_i)}{P(X)}
$$

- The class with the **highest posterior probability** is chosen:
  
$$
\hat{C} = \arg\max_{C_i} P(C_i | X)
$$

---

3. **Why Denominator is Ignored**

Since \(P(X)\) is the same for all classes:

$$
P(C_i | X) \propto P(X | C_i) \, P(C_i)
$$

Thus, we only compare numerators across classes.

---

4. **Naïve Assumption**

Naïve Bayes assumes **features are conditionally independent** given the class:

$$
P(X | C_i) = \prod_{j=1}^n P(x_j | C_i)
$$

This simplifies computation drastically.

---

5. **Final Decision Rule**

$$
\hat{C} = \arg\max_{C_i} \, P(C_i) \prod_{j=1}^n P(x_j | C_i)
$$

---

✅ The method is called **Naïve Bayes** because:
- It applies Bayes theorem,  
- It makes the "naïve" assumption of conditional independence of features.


# Naive Bayes

Let  
$$
X = \{x_1, x_2, x_3, \dots, x_n\}
$$  
$$
C_k = \{c_1, c_2, c_3, \dots, c_k\}
$$  

---

### Bayes Rule
$$
P(C_k | X) = \frac{P(X | C_k) P(C_k)}{P(X)}
$$

---

### Expanding with Chain Rule for Conditional Terms
$$
P(C_k | X) \propto P(X, C_k)
$$

$$
P(X, C_k) = P(x_1, x_2, x_3, \dots, x_n, C_k)
$$

Using the chain rule:
$$
P(x_1, x_2, x_3, \dots, x_n, C_k) = P(x_1 | x_2, x_3, \dots, x_n, C_k) \cdot P(x_2, x_3, \dots, x_n, C_k)
$$

So,
$$
P(C_k | X) = a \cdot P(x_2, x_3, \dots, x_n, C_k)
$$

---

### Recursive Expansion
$$
P(x_2, x_3, \dots, x_n, C_k) = P(x_2 | x_3, x_4, \dots, x_n, C_k) \cdot P(x_3, x_4, \dots, x_n, C_k)
$$

$$
= P(x_1 | x_2, x_3, \dots, x_n, C_k) \cdot P(x_2 | x_3, x_4, \dots, x_n, C_k) \cdot \dots \cdot P(x_n | C_k) \cdot P(C_k)
$$

---

### Conditional Independence Assumption
$$
P(A | B) = P(A)
$$

$$
P(A | B, C) = P(A | C)
$$

So,
$$
P(C_k | X) \propto P(C_k) \prod_{i=1}^n P(x_i | C_k)
$$
