### **STACKING AND BLENDING ENSEMBLE**
- Stacking is one of the best solution to use in for the data.
- **Steps**
    * Train your base models.
    * Run all three models on the same models
        * Then create a new dataset from the predictions to train the ```meta model ```. 
    * The input for the meta model will be the predictions and the target column is the same.

* It is different from boosting and baggign as it can have different base models. In bagging and boosting we need to have the same base models.
- In case of bagging or bossting we take the masjority or assign the weightage to the models and take the sum.

**The problem with the stacking is that**
- We are training in the same data and making and predicting on the same data.
- This will lead to overfitting to the ```meta model```.
\
**The method we use is ```Hold out method``` and ```K fold method```**:
- **Hold out method**:Training the base models and meta model with the different part of the data. It is called blending.
    * We first break the data in to DTrain and DTest. Then the train data is again divided in D_Train, D_validation.
    D_Train is used to train base models. Then D_validation is used to for the prediction. Then the prediction of each of the models will be given to the meta model.
-**k fold method** : Use of this method is called stacking. 
    * We divide the dataset in the training and the testing set. Then training set is divided in folds with value of k different according to the need. The k-1 folds are used for to train the base models. and 1 for the prediction. Then again the another k folds are created and the base models are train thus to make the total no of preidiction equal to the no of the training set. Thus base models*k-folds are trained during the training process. Then on the pridiction data is used to train the meta model. The whole training is used to train the base models. That is first the meta model is trained then again D_test dataset is used to test the model. 

**Multi level stacking**
- We make the complex architecture by layring the base models then sending the prediction to the layer forward and then finally combing the predictions to get the more accurate prediction. 
- Here the trainig set is divided to the multiple training set. Using a single training data the base models are trained. Then the prediction are obtained from the next training set. Them the preediction are sent to the next layer containing the multiple base models. Then another training set yet unseen is given to predict for this layer base models. Then the prediction is obtained which is later given to the meta model. 
- We have different number of the base models in each layer. This layer and multiple stacking is the very good idea to obtaing the better model.

````python
import numpy as np
import pandas as pd
from sklearn.ensemble import StackingClassifier
````

````python
df=pd.read_csv('C:/Users/Acer/OneDrive/Documents/MACLEARNING/machinelearning/data/heart.csv')
df.head()
````

````python
X=df.drop(columns=['target'])
y=df['target']
````

````python
X
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
````

````python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
````

````python
estimators=[
    ('rf',RandomForestClassifier(n_estimators=10,random_state=42)),
    ('knn',KNeighborsClassifier(n_neighbors=10)),
    ('gbt',GradientBoostingClassifier(n_estimators=10))
]
````

````python
clf=StackingClassifier(estimators=estimators,final_estimator=LogisticRegression(),cv=10)
````

````python
clf.fit(X_train,y_train)
````

````python
from sklearn.metrics import accuracy_score
y_pred=clf.predict(X_test)
````

````python
accuracy_score(y_test,y_pred)
````

````python
log=LogisticRegression()
log.fit(X_train,y_train)
y_pred2=log.predict(X_test)
accuracy_score(y_test,y_pred2)
````
